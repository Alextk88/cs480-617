```python
sql = '''
SELECT "DISTRICT",
       "OCCURRED_ON_DATE",
       "SHOOTING",
       "OFFENSE_DESCRIPTION"
FROM "b973d8cb-eeb2-4e7e-99da-c92938efc9c0"
WHERE CAST("SHOOTING" AS INT)=1
    '''
```

```python
bpd_logs = "https://data.boston.gov/api/3/action/datastore_search_sql"
```

```python
import urllib, requests
```

```python
url = bpd_logs+"?sql="+urllib.parse.quote(sql)
rows = requests.get( url ).json()

shootings = rows['result']['records']
```

```python
import requests
import pandas as pd

# API endpoint
url = "https://data.boston.gov/api/3/action/datastore_search"

params = {
    "resource_id": "b973d8cb-eeb2-4e7e-99da-c92938efc9c0",
    "limit": 50000
}

# request data
response = requests.get(url, params=params)
data = response.json()

# convert to dataframe
records = data["result"]["records"]
df = pd.DataFrame(records)

# preview
df.head()
```

```python
import pandas as pd

# Chart 1: Top crimes (overall)
crime_counts = (
    df["OFFENSE_DESCRIPTION"]
    .value_counts()
    .head(8)
    .reset_index()
)
crime_counts.columns = ["OFFENSE_DESCRIPTION", "count"]

# Chart 2: Shoplifting by district
shoplifting_df = df[df["OFFENSE_DESCRIPTION"] == "LARCENY SHOPLIFTING"]

district_counts = (
    shoplifting_df["DISTRICT"]
    .value_counts()
    .reset_index()
)
district_counts.columns = ["DISTRICT", "count"]

crime_counts, district_counts
```

```python
import json

crime_json = crime_counts.to_dict(orient="records")
district_json = district_counts.to_dict(orient="records")

print(json.dumps(crime_json))
print(json.dumps(district_json))
```
